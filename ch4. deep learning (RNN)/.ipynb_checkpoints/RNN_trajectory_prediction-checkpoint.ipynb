# RNN for vehicle trajectory prediction


**Trajectory prediction** is a key module in autonomous driving motion planning, it is crucial in making safe decisions. In this tutorial, a RNN based method is used to predict the vehicles' trajectory under a crash immenent situation.

## 1. Settings
As shown in the following GIF, the trajectoy data is collected within this scenario where there are 2 types of vehicles: human driven vehicles (HDVs) and connected autonomous vehicles (HDVs). In the figure, red vehicle is the CAV while the rest are all HDVs. 

### Notations:
This scenario is created when one HDV (yellow) vehicle is making a aggressive lane changing when the CAV (red) is in its blind spot. We denote this yellow HDV as  lane-changing HDV **LHDV**. The rest vehicles are the grey vehicle in the downstream of LHDV, denoted as front HDV **FHDV**, the grey vehicle behind CAV, denoted as behind HDV **BHDV**.

In this situation, if without proper escaping maneuver, there will be a collision between CAV and LHDV, and this collision may even propagate to more collisions between CAV and BHDV.

Therefore, in order to avoid the collision between each pair of vehicles, we need to predict the potential crash point (time and location) before taking any decisions. However, due to the  

2 trajectory prediction models for HDV and CAV are needed when predicting the potential crashes.

## 3. Prepare the dataset

```python
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
import torch 

# DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# if multiple gpu, can also specify the specific gpu
DEVICE = "cuda:1"
```

```python
class TrajectoryDataset(Dataset):
    def __init__(self, file_name, window_size=5, cav_flag=False):
        self.raw_data = pd.read_csv(file_name)
        self.window_size = window_size
        if cav_flag:
            self.x_columns = ['px','py','sx','sy','ax','ay','throttle','steer','brake']
        else:
            self.x_columns = ['px','py','sx','sy','ax','ay']
        self.y_columns = ['px','py','sx','sy','ax','ay']

        self.X, self.Y = self.preprocessing_data(self.raw_data, self.window_size, self.x_columns, self.y_columns)

    def preprocessing_data(self, data_frame, window_size, x_columns, y_columns): 
        data_frame = data_frame.copy()
        X = []
        Y = []
        for (veh_id, episode), trajectories in data_frame.groupby(['veh_id','episode']):# iterate over groupby object
            trajectories.sort_values(by='episode_step',inplace=True)
            # Select the columns and zero pad at the beginning 
            zero_padded = np.vstack([np.zeros([window_size-1,len(x_columns)]),np.array(trajectories[x_columns])])
            sequences = [zero_padded[i:i+window_size] for i in range(len(trajectories))]
            xs = sequences[:-1] 
            ys = sequences[1:] # one timestep later than xs 
            ys = [y[:,:len(y_columns)] for y in ys] # Truncate the control details since they are not predictable in ys
            X.extend(xs)
            Y.extend(ys)
        return X,Y

    def __len__(self):
        return len(self.X)

    def __getitem__(self,idx):
        if torch.is_tensor(idx):
            idx = idx.to_list()
        return torch.tensor(self.X[idx]).float().to(DEVICE), torch.tensor(self.Y[idx]).float().to(DEVICE) # Tensorfy and put on the GPU/CPU
```

```python
CAV_traj_dataset = TrajectoryDataset('./cav_info.csv', window_size=5, cav_flag=True)
HDV_traj_dataset = TrajectoryDataset('./hdv_info.csv', window_size=5, cav_flag=False)

len(CAV_traj_dataset)
```

## 4. Build RNN model

```python
import torch.nn as nn
```

```python
# Model class
class RNN_Predictor(nn.Module):
    def __init__(self,input_dim, output_dim, encode_dim=None, rnn_hidden_dim=16):
        super().__init__()
        if encode_dim:
            self.encoder = nn.Sequential(
                nn.Linear(input_dim, encode_dim),
                nn.ELU())
        else:
            self.encoder = nn.Identity()
            encode_dim = input_dim
        self.rnn = nn.GRU(input_size=encode_dim, hidden_size=rnn_hidden_dim, batch_first=True)
        self.output_layer = nn.Linear(in_features=rnn_hidden_dim, out_features=output_dim)
    
    def forward(self, input_batch): 
        # input_batch shape: batch_size * seq_len * feature_size
        embeddings = self.encoder(input_batch) # batch_size * seq_len * encode_dim
        hidden, _ = self.rnn(embeddings) # batch_size * seq_len * hidden_dim
        output = self.output_layer(hidden) # batch_size * seq_len * hidden_dim
        return output
```

```python
def train_one_epoch(model, optimizer, train_data_loader, valid_data_loader=None):
    model = model.to(DEVICE)
    loss_fn = nn.MSELoss().to(DEVICE)
    batch_id = 0
    training_loss = 0
    for x_batch, y_batch in train_data_loader:
        batch_id += 1
        # forward pass
        outputs = model.forward(x_batch)
        loss = loss_fn(outputs,y_batch)
        training_loss += loss.item()
        # back propagate
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # print statistics
        if batch_id%100 == 0:
            print("Batch number {}: ------  loss: {}".format(batch_id,loss.item()))

    training_loss /= batch_id
    # after each epoch, run through the validation dataset (if any)
    valid_loss = None
    if valid_data_loader:
        batch_id = 0
        valid_loss = 0
        for x_batch, y_batch in valid_data_loader:
            batch_id += 1
            outputs = model.forward(x_batch)
            loss = loss_fn(outputs,y_batch)
            valid_loss += loss.item()
        valid_loss /= batch_id # devide by each batch
    return training_loss, valid_loss
```

### CAV trajectory prediction model

```python
CAV_predictor = RNN_Predictor(input_dim=9, encode_dim=16, output_dim=6)
CAV_optimizer = torch.optim.Adam(CAV_predictor.parameters() ,lr=0.001)
CAV_train_data_loader = DataLoader(CAV_traj_dataset, batch_size=32, shuffle=True)
```

```python
training_losses = []
for i in range(1000):
    if i%100 ==0: 
        print(" ------ start epoch %d ------"%(i+1))
    training_loss, valid_loss = train_one_epoch(CAV_predictor, CAV_optimizer, CAV_train_data_loader)
    training_losses.append(training_loss)
```

```python
plt.plot(training_losses)
plt.xlabel('epoch')
plt.ylabel('training loss')
plt.title('CAV loss training curve')
```

```python
cav_model_path = './models/CAV_predictor.pt'
torch.save(CAV_predictor.state_dict(),cav_model_path)
```

### HDV trajectory prediction model

```python
HDV_predictor = RNN_Predictor(input_dim=6,encode_dim=16, output_dim=6) # Since HDV has no control command, only 6 as input feature vec
HDV_optimizer = torch.optim.Adam(HDV_predictor.parameters() ,lr=0.001)
HDV_train_data_loader = DataLoader(HDV_traj_dataset, batch_size=64, shuffle=True)
```

```python
training_losses = []
for i in range(1000):
    if i%100 ==0: 
        print(" ------ start epoch %d ------"%(i+1))
    training_loss, valid_loss = train_one_epoch(HDV_predictor, HDV_optimizer, HDV_train_data_loader)
    training_losses.append(training_loss)
```

```python
# Save HDV_predictor
hdv_model_path = './models/HDV_predictor.pt'
torch.save(HDV_predictor.state_dict(),hdv_model_path)
```

```python
plt.plot(training_losses)
plt.xlabel('epoch')
plt.ylabel('training loss')
plt.title('HDV loss training curve')
```

## 5. Test the model

```python
# load the model model.
CAV_predictor = RNN_Predictor(input_dim=9, encode_dim=16, output_dim=6) # build a new model
CAV_predictor.load_state_dict(torch.load(cav_model_path)) # load the trained weights
CAV_predictor.to(DEVICE) # make it on the GPU

# test the model
CAV_predictor.eval()
random_x, random_y = CAV_traj_dataset[np.random.randint(len(CAV_traj_dataset))]
output = CAV_predictor.forward(random_x.view([1,5,9])).view([5,6]).detach().cpu().numpy() # plot in numpy.ndarray, move CUDA tensor to CPU, then transfer to numpy
plt.scatter(output[:,0], output[:,1],label='prediction',marker='o')     
plt.scatter(random_y[:,0].cpu().numpy(),random_y[:,1].cpu().numpy(),label='ground truth',marker='*')
plt.legend()
```

```python
# load the model model.
HDV_predictor = RNN_Predictor(input_dim=6, encode_dim=16, output_dim=6) # build a new model
HDV_predictor.load_state_dict(torch.load(hdv_model_path)) # load the trained weights
HDV_predictor.to(DEVICE) # make it on the GPU

# test the model
HDV_predictor.eval()
random_x, random_y = HDV_traj_dataset[np.random.randint(len(HDV_traj_dataset))]
output = HDV_predictor.forward(random_x.view([1,5,6])).view([5,6]).detach().cpu().numpy()
plt.scatter(output[:,0], output[:,1], label='prediction',marker='o')
plt.scatter(random_y[:,0].cpu().numpy(),random_y[:,1].cpu().numpy(),label='ground truth',marker='*')
plt.legend()
```
